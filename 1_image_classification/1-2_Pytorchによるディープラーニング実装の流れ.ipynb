### PyTorchによるディープラーニング実装の流れ
1. 前処理、後処理、そしてネットワークモデルの入出力を確認
2. Datasetの作成
3. DataLoaderの作成
4. ネットワークモデルの作成
5. 順伝搬(forward)の定義
6. 損失関数の定義
7. 最適化手法の設定
8. 学習・検証の実施
9. テストデータで推論

まずこれから実装するディープラーニングの応用手法の全体像を把握する。つまり、前処理、後処理、そしてネットワークモデルの入出力を確認する。

次にクラスDatasetを作成する。Datasetとは入力するデータとそのラベルなどをペアにして保持したクラス。Datasetにはデータに対する前処理クラスのインスタンスを与え、対象データのファイルを読み込む際に、前処理を自動で適用するように設定する。Datasetは訓練データ、検証データ（そしてテストデータ）、それぞれについて作成する。

次にクラスDataLoaderを作成する。DataLoaderはDatasetからどのようにデータを取り出すのかを設定するクラス。ディープラーニングではミニバッチ学習を行い、複数のデータを同時にDatasetから取り出してネットワークを学習させることが一般的。DataLoaderはDatasetからミニバッチを取り出しやすくするためのクラス。訓練データ、検証データ（そしてテストデータ）それぞれのDataLoaderを作成する。DataLoaderが完成すれば入力データに対する事前準備は完了となる。

次にネットワークモデルを作成する。ネットワークモデルの作成はゼロから全て自分で作成するケース、学習済みモデルをロードして使用するケース、学習済みモデルをベースに自分で改変するケースがあります。ディープラーニングの応用手法では、学習済みモデルをベースに自分で改変する場合が多い。

ネットワークモデルの作成後は、ネットワークモデルの順伝搬関数(forward)を定義する。ネットワークモデルが単純な場合は、データはモデルを構築している層を前から後に流れるだけ。しかし、ディープラーニングの応用手法は順伝搬が複雑な場合が多い。例えば、ネットワークが途中で分岐したりする。このような複雑な順伝搬を実現するためには、順伝搬関数(forward)をきちんと定義する。

順伝搬関数が定義できれば、誤差逆伝播(backpropagation)をするための損失関数を定義する。簡単なディープラーニング手法であれば損失関数は2乗誤差など単純な関数だが、ディープラーニングの応用手法においては、損失関数が非常に複雑な形をしているケースもある。

次にネットワークモデルの結合パラメータを学習させる際の最適化手法を設定する。誤差逆伝播によって結合パラメータの誤差に対する勾配が求まる。最適化手法では、この勾配を使用し、結合パラメータの修正量をどのように計算するのか設定する。最適化手法としては例えばMomentum SGDなどが使用される。

以上でディープラーニングの学習のための設定は完了。続いて、学習と検証を行う。基本的にはepochごとに訓練データでの性能と検証データでの性能を確認し、検証データの性能が向上しなくなったら、その後は訓練データに対して過学習に陥っていくため、そのタイミングで学習を終了させることが多い。このような検証データの性能が向上しなくなったら学習を終了する手法をearly stoppingと呼ぶ。

学習が完了すれば、最後にテストデータに対して推論を行う。
